### Condicionais (if, else, elif)

Em pyhton, existe uma instrução de seleção de fluxo chamada `if`. Com ela podemos escolher o fluxo de execução baseado em uma `condição lógica`.

As instrução `else` pode ser combinada com um `if`, que indica o fluxo contrário da condição lógica.

E a instrução `elif` permite que você adicione mais de um fluxo condicional em sequência.

```python
# Declarando as variáveis
nota_final_aluno = 7.5  # nodta final: [0, 10]
porcentagem_frequencia = 60/100 # frequência: [0, 1]
```

```python
# Instrução IF
if nota_final_aluno >= 5 and porcentagem_frequencia > 70/100:
  print("aluno aprovado")

# Neste caso, os alunos com nota final igual ou superiores a 5 e com pelo menos 
# 70% de frequência serão aprovados
```

```python
# Instrução IF + ELSE
if nota_final_aluno >= 5 and porcentagem_frequencia > 70/100:
  print("aluno aprovado")
else:
  print("aluno reprovado")

# Neste caso, os alunos com nota final igual ou superiores a 5 e com pelo menos 
# 70% de frequência serão aprovados, e os demais, reprovados
```

```python
# Instrução IF + ELIF + ELSE
if nota_final_aluno >= 5 and porcentagem_frequencia > 70/100:
  print("aluno aprovado")
elif nota_final_aluno < 5 and porcentagem_frequencia > 70/100:
  print("aluno em recuperação")
else:
  print("aluno reprovado")

# Neste caso, os alunos com nota final igual ou superiores a 5 e com pelo menos 
# 70% de frequência serão aprovados.
# Caso o aluno tenha nota menor que 5, mas pelo menos 70% da frequência, então o
# aluno está em recuperação.
# Caso contrário, o aluno é reprovado
```

### Loops de repetição

Loops de repetição são usados para `repetir um bloco de código`. Essa repetição pode ser finita ou infinita, tudo depende da lógica que for implementada.

Dentre os loops, usaremos 2 tipos:

* `for` - Usado para repetições finitas. Usamos quando temos certeza de quantas vezes a iteração irá se repetir.
* `while` - Pode ser usado para repetições finitas ou infinitas. Funciona a partir de uma `condição lógica`.

**Observação**: Sempre que utilizar o `while`, tome muito cuidado com a `condição lógica` que for implementada. Uma condição lógica errada, pode levar o seu progama a entrar num estado em que a `repetição do loop jamais irá teminar`, a menos que o programa seja interrompido externamente.

```python
# Exemplo do loop FOR
total_repeticoes = 10
for i in range(total_repeticoes):
  print(i)

# Utilizando a função range(x), estamos criando uma sequência de números, 
# começando pelo 0, x vezes.
# Se começarmos pelo 0 e iterarmos x vezes, sendo o valor de x = 10, então 
# teremos uma lista de 0 até 9, totalizando 10 números.
```

```python
# Exemplo do loop FOR
x = 0
while x < 10:
  print("O valor de x é:", x)
  x = x + 1

# Com o while, utilizamos a condição lógica x < 10
# Enquanto x não for menor que 10, a iteração continua.
```

### Exercício 1

Implementar um programa que imprima todos os `números ímpares` de 1 a 100.

### Gabarito do Exercício 1

```python
# Realizar um loop de 1 até 100 para percorrermos todos os números de 1 até 100
# Nesse comando `for`, já estamos declarando a variável `numero_atual` que irá
# acumular o valor atual para nós
for numero_atual in range(1, 100+1):

  # Obtendo o resto da divisão
  resto_divisao = numero_atual % 2
  
  # Verificando se o número é impar
  # Nesse caso, quando o resto da divisão por 2 é 1, quer dizer que o número é 
  # ímpar
  if resto_divisao == 1:
    # Imprimindo o número atual que é ímpar
    print(numero_atual)
```

### Exercício 2

Implementar um programa que imprima todos os números de 1 a 100 que `são divisíveis por 3` e que `são pares`.

### Gabarito do exercício 2

```python
# Realizar um loop de 1 até 100 para percorrermos todos os números de 1 até 100
# Nesse comando `for`, já estamos declarando a variável `numero_atual` que irá
# acumular o valor atual para nós
for numero_atual in range(1, 100+1):

  # Obtendo o resto da divisão
  resto_divisao_por_3 = numero_atual % 3
  resto_divisao_por_2 = numero_atual % 2

  # Verificando se o número é divisível por 3 e se é par
  # Quando o resto da divisão por 3 é 0, quer dizer que o número é 
  # divisível por 3
  # Quando o resto da divisão por 2 é 0, quer dizer que o número é par
  if resto_divisao_por_3 == 0 and resto_divisao_por_2 == 0:
    # Imprimindo o número atual que é divisível por 3 e é par
    print(numero_atual)
```
